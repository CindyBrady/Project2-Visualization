```python
import pandas as pd
```

```python
df = pd.read_csv("outdoordata.csv")
df = df[["Store Name", "Lat, Long", "Country", "City", "Revenue"]].copy()
df
```

```python
rev_total = df.groupby(["Store Name", "Lat, Long", "Country", "City"]).sum()
rev_df = pd.DataFrame(rev_total).reset_index()
rev_df
```

```python
rev_df.to_csv("mapdata.csv", index = False)
```
